```python
# requests
import requests
response = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import json
data = json.loads(response.text)

print('ä¸€æ¬¡å›å‚³{}ç­†'.format(len(data)))
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«ä»¥ä¸‹é€™äº›æ¬„ä½:')
#data[0] # ç¬¬ä¸€ç­†è³‡æ–™
for i in data[0]:
    print(i)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€


alist = []
for i in data:
    dic = {}
    dic['title'] = i['title']
    dic['createdAt'] = i['createdAt']
    dic['commentCount'] = i['commentCount']
    dic['likeCount'] = i['likeCount']
    alist.append(dic)
    
print(alist)    
```

```python
# 3. è¨ˆç®—ç†±é–€çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
average_comment = 0
average_like = 0

for i in data:
    average_comment += i['commentCount']
    average_like += i['likeCount']
print('ç†±é–€å¹³å‡ç•™è¨€äººæ•¸{}'.format(average_comment/len(data)))
print('ç†±é–€å¹³å‡æŒ‰è®šäººæ•¸{}'.format(average_like/len(data)))
average_comment
```

```python
# 3. è¨ˆç®—éç†±é–€ç†±é–€çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
response_false = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data_false = json.loads(response_false.text)

not_average_comment = 0
not_average_like = 0

for i in data_false:
    not_average_comment += i['commentCount']
    not_average_like += i['likeCount']
print('éç†±é–€å¹³å‡ç•™è¨€äººæ•¸{}'.format(not_average_comment/len(data_false)))
print('éç†±é–€å¹³å‡æŒ‰è®šäººæ•¸{}'.format(not_average_like/len(data_false)))
not_average_comment
```
